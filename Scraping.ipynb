```python
# Import Statements
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url="https://in.tradingview.com/markets/stocks-india/market-movers-large-cap/"
```

```python
# Checking the response of url
page=requests.get(url)
page
```

```python
# Using Beautiful soup library for web scraping 
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find("table")
table=soup.find("tbody")
table
```

```python
rows=table.find_all("tr")
rows
```

```python
stocks_info=[]

for i in rows:
    cols=i.find_all("td")
    #print(cols)
    market_stocks=[C.text for C in cols]
    #print(market_stocks)
    stocks_info.append(market_stocks)
print(stocks_info)
```
